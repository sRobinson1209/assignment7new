```python
import matplotlib.pyplot as plt
import networkx as nx

def draw_graph(array):
    # Initialize directed graph
    G = nx.DiGraph()

    n = len(array)

    # Add vertices to the graph
    for i, (v, x) in enumerate(array):
        G.add_node(v)

    # Add edges to the graph
    for v, x in array:
        right_vertex = array[(i + x) % n][0]  # Vertex x positions to the right
        left_vertex = array[(i - x) % n][0]   # Vertex x positions to the left
        G.add_edge(v, right_vertex)
        G.add_edge(v, left_vertex)

    # Draw the graph
    pos = nx.circular_layout(G)  # Circular layout for circular array
    nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=12, font_weight='bold', arrowsize=20)
    plt.title("Directed Graph")
    plt.show()

if __name__ == "__main__":
    # Example input array
    input_array = [('I', 2) , ('A', 5) , ('E', 4) , ('F',1) , ('T', 2) , ('S', 3)]

    draw_graph(input_array)
```
